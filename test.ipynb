```python
from git import Repo

from Objects.Version import Version
from projet.leo.get_versions import get_versions
from projet.leo.metrics.metric_02_removed_lines import metric_02_removed_lines
from projet.leo.metrics.metric_01_added_lines import metric_01_added_lines
```

```python


repository: Repo = Repo("data/hive_data/hiveRepo/")

versions: [Version] = get_versions(repository)
# diff_between_2_versions(versions[-2], versions[-1])
```

```python
print(len(versions[-1].getDiff().keys()))
```

```python
from projet.leo.metrics.metric_01_added_lines import metric_01_added_lines
files = list(Version.files_found)
i = 0
for file in files:
    if i % 10 == 0:
        print(i)
    i += 1
    print(file, metric_01_added_lines(file, versions[-1]))
```

```python
from projet.leo.metrics.metric_03_commit_count import metric_03_commit_count
m = 0
f = None
for file in files:
    if metric_03_commit_count(file, versions[-2]) != 0:
        print(file, metric_03_commit_count(file, versions[-2]))
        if metric_03_commit_count(file, versions[-2]) > m:
            m = metric_03_commit_count(file, versions[-2])
            f = file
print(m, file)
```

```python
from projet.leo.metrics.metric_06_developer_count import metric_06_developer_count
m = 0
for file in files:
    k = metric_06_developer_count(file, versions[-1])
    if file.endswith(".java") and k > m:
        print(k, file)
        m = k
        
```
